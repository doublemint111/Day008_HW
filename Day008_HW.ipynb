```python
import requests
# 引入函式庫
r = requests.get('https://github.com/timeline.json')
# 想要爬資料的目標網址
response = r.text
# 模擬發送請求的動作
```

```python
print(response)
response['message']
```

```python
print(type(response))

# print(response['message'])
print(response[:100])
```

```python
import json
response = json.loads(response)

print(type(response))
print(response['message'])
```
